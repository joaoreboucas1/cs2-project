# CMB+LSS chains

```python
from importlib import reload
from copy import copy

import numpy as np
from getdist import plots as gdplot

import utils
```

```python
# Load chains
settings = copy(utils.DEFAULT_GETDIST_SETTINGS)
settings['smooth_scale_2D'] = 0.4
indices = [40, 41, 42]
chains = {}
for i in indices:
    chains[i] = utils.load_chain(i, settings=settings)
```

```python
p = gdplot.get_subplot_plotter()
p.settings.axes_fontsize=25
p.settings.axes_labelsize=30
p.settings.legend_fontsize=30
p.triangle_plot(
    [chains[42], chains[40], chains[41]],
    ["omegam", "S8", "w", "wa", "H0", "cs2_0"],
    filled=[True, False, False],
    contour_colors=utils.colors,
    contour_lws=3,
    contour_ls=["-", "--", "-."],
    legend_labels=["$c_s^2$ free", "$c_s^2 = 1.0$", "$c_s^2 = 0.0$"],
    legend_loc="upper right"
)
p.fig.suptitle("Data: DES-Y3 Shear + Planck PR4 LIPOP & Lensing + DESI DR2 BAO + PantheonPlus SN + eBOSS RSD", fontsize=20, y=1.02)
p.export("plots/triangle_desshear_cmb_bao_sn.pdf")
```
