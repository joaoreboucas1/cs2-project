# Considering MG chains with LCDM prior, constant $c_s^2$

```python
import numpy as np
import getdist
from getdist import plots

import utils
```

```python
indices = list(range(60, 64)) + [49]
chains = {}
for i in indices:
    chains[i] = utils.load_chain(i, burn_in=0.2, smooth_2D=0.3)
```

```python
p = plots.get_subplot_plotter()
p.settings.axes_fontsize=25
p.settings.axes_labelsize=30
p.settings.legend_fontsize=27
p.triangle_plot(
    [chains[60], chains[62], chains[63], chains[49]],
    ["H0", "omegam", "S8", "cs2_0", "alpha_K_0"],
    filled=[False, False, False, True],
    contour_colors=utils.colors[:3] + ["gray"],
    contour_lws=3,
    contour_ls=["-", "--", "-.", "-"],
    legend_labels=["$\\alpha_K \\propto \\Omega_\\mathrm{DE}$", "$\\alpha_K \\propto \\alpha_B$", "Cubic Galileon Self-Accelerating", "Fluid"],
    legend_loc="upper right",
    param_limits={"H0": [67.3, 69.2], "omegam": [0.29, 0.315]}
)
p.fig.text(0.3, 1.01, "Data: Planck + DESI + DES-Dovekie", fontsize=20)
p.fig.savefig("plots/triangle_ds1_subluminal_mg_lcdm.pdf", bbox_inches="tight")
```
