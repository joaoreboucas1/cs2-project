# Triangle plots for subluminal chains, dyamical $c_s^2$

```python
import numpy as np
import getdist
from getdist import plots

import utils
```

```python
indices = list(range(68, 72)) + [48]
chains = {}
for i in indices:
    chains[i] = utils.load_chain(i, burn_in=0.2, smooth_2D=0.3)
```

```python
p = plots.get_subplot_plotter()
p.settings.axes_fontsize=25
p.settings.axes_labelsize=30
p.settings.legend_fontsize=30
p.triangle_plot(
    [chains[68], chains[70], chains[71], chains[48]],
    ["H0", "omegam", "S8", "w", "wa", "cs2_0", "cs2_a"],
    filled=[False, False, False, True],
    contour_colors=utils.colors[:3] + ["gray"],
    contour_lws=3,
    contour_ls=["-", "--", "-.", "-"],
    legend_labels=["$\\alpha_K \\propto \\Omega_\\mathrm{DE}$", "$\\alpha_K \\propto \\alpha_B$", "Cubic Galileon Self-Accelerating", "Fluid"],
    legend_loc="upper right",
    param_limits={"w": [-1.01, -0.55], "wa": [-1.55, 0.1], "H0": [65, 69], "omegam": [0.295, 0.34]}
)
p.fig.text(0.3, 1.01, "Data: Planck + DESI + DES-Dovekie", fontsize=20)
p.fig.savefig("plots/triangle_ds1_subluminal_dyncs2_mg.pdf", bbox_inches="tight")
```

```python
p = plots.get_single_plotter()
p.settings.axes_fontsize=25-10
p.settings.axes_labelsize=30-10
p.settings.legend_fontsize=30-10
p.plot_2d(
    [chains[59], chains[48]],
    ["w", "wa"],
    filled=[True, False],
    colors=[utils.colors[0], "black"],
    lws=2,
    ls=["-", "--", "-.", "-"],
    labels=["Cubic Galileon", "Fluid"],
    legend_loc="upper right",
    lims=[-1.01, -0.55, -2, 0.1]
)
p.add_x_marker(-1)
p.add_y_marker(0)
p.add_legend(["Cubic Galileon", "Fluid"], fontsize=15)
# p.fig.text(0.17, 1.01, "Data: Planck PR4 + DESI DR2 BAO + DES-Dovekie SN", fontsize=15)
p.fig.savefig("plots/triangle_ds1_subluminal_dyncs2_mg_w0waonly.pdf", bbox_inches="tight")
```
