```python
import numpy as np
import getdist
from getdist import plots

import utils
```

```python
indices = [45, 46, 48, 49]
chains = {}
for i in indices:
    chains[i] = utils.load_chain(i)
```

```python
p = plots.get_subplot_plotter()
p.settings.axes_fontsize=25
p.settings.axes_labelsize=30
p.settings.legend_fontsize=30
p.triangle_plot(
    [chains[45], chains[46], chains[48]],
    ["omegam", "S8", "w", "wa", "cs2_0"],
    filled=[True, False, False],
    contour_colors=utils.colors,
    contour_lws=3,
    contour_ls=["-", "--", "-."],
    legend_labels=["$\\alpha_K$ constant", "$\\alpha_K \\propto \\Omega_\\mathrm{DE}$", "Fluid"],
    legend_loc="upper right"
)
p.fig.text(0.3, 1.01, "Data: Planck + DESI + DES-Dovekie", fontsize=20)
p.fig.savefig("plots/cs2mg_test.pdf", bbox_inches="tight")
```
