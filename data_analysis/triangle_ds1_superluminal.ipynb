# Triangle plots for superluminal chains, constant $c_s^2$

```python
import numpy as np
import getdist
from getdist import plots

import utils
```

```python
indices = list(range(64, 68)) + [48]
chains = {}
for i in indices:
    chains[i] = utils.load_chain(i, burn_in=0.2, smooth_2D=0.3)
```

```python
p = plots.get_subplot_plotter()
p.settings.axes_fontsize=25
p.settings.axes_labelsize=30
p.settings.legend_fontsize=30
p.triangle_plot(
    [chains[64], chains[66], chains[67], chains[48]],
    ["H0", "omegam", "S8", "w", "wa"],
    filled=[False, False, False, True],
    contour_colors=utils.colors[:3] + ["gray"],
    contour_lws=3,
    contour_ls=["-", "--", "-.", "-"],
    legend_labels=["$\\alpha_K \\propto \\Omega_\\mathrm{DE}$", "$\\alpha_K \\propto \\alpha_B$", "Cubic Galileon-like", "PPF"],
    legend_loc="upper right",
    param_limits={"w": [-1.01, -0.55], "wa": [-1.55, 0.1], "H0": [65, 69], "omegam": [0.295, 0.34]}
)
p.fig.text(0.3, 1.01, "Data: Planck + DESI + DES-Dovekie", fontsize=20)
p.fig.savefig("plots/triangle_ds1_superluminal_mg.pdf", bbox_inches="tight")
```

```python
p = plots.get_subplot_plotter(width_inch=9, subplot_size_ratio=3/4)
p.settings.axes_fontsize=20
p.settings.axes_labelsize=20
p.settings.legend_fontsize=15
p.plots_1d(
    [chains[64], chains[65], chains[66], chains[67]],
    ["cs2_0", "alpha_K_0"],
    colors=utils.colors,
    lws=3,
    ls=["-", "--", "-.", "-"],
    nx=2,
    legend_labels=["$\\alpha_K \\propto \\Omega_\\mathrm{DE}$", "K-essence-like", "$\\alpha_K \\propto \\alpha_B$", "Cubic Galileon-like"],
    legend_loc="upper right",
    legend_ncol=4,
    param_limits={"w": [-1.01, -0.55], "wa": [-1.55, 0.1], "H0": [65, 69], "omegam": [0.295, 0.34]}
)
p.fig.savefig("plots/lambda_mg_cs2_superluminal_ds1.pdf", bbox_inches="tight")
```
